```python
import pandas as pd
import numpy as np
```

```python
names = ['Minute Index', 'Landing Page', 'Source', 'Device Category', 'bounces']
source = pd.read_excel('source.xlsx', header=None, names=names)
```

```python
source.head()
```

```python
source = source.drop([0])
```

```python
source.head()
```

```python
names1 = ['minute_index', 'page', 'source', 'device_type', 'bounces']
count_pos = pd.read_excel('count_pos.xlsx')
```

```python
count_pos.head()
```

```python
names1 = ['minute_index', 'page', 'source', 'device_type', 'bounces']
sour_pos = pd.read_excel('sour_pos.xlsx')
```

```python
sour_pos.head()
```

```python
sour_pos = sour_pos.drop(['Bounces','Bounce Rate','Session Duration'], axis=1)
```

```python
new_pos = pd.merge(count_pos, sour_pos,  how='inner', left_on=['Minute Index','Landing Page'], right_on = ['Minute Index','Landing Page'])
```

```python
new_pos
```

```python
country = pd.read_excel('country.xlsx')
```

```python
country.head()
```

```python
new_df = pd.merge(country, source,  how='inner', left_on=['Minute Index','Landing Page'], right_on = ['Minute Index','Landing Page'])
```

```python
new_df = new_df.drop('bounces', axis=1)
```

```python
new_df
```

```python
df_concat = pd.concat([new_df,new_pos])
```

```python
df_concat
```

```python
df_concat['clas'] = np.where(((df_concat['Bounce Rate']>0)&(df_concat['Bounce Rate']<1))&(df_concat['Session Duration']>=8) , '1', '0')
```

```python
final = df_concat.drop('Minute Index',axis=1)
```

```python
final
```

```python
new = final[final['clas']=='1']
new
```
